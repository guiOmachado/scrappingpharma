```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import math
import numpy as np

termo_de_pesquisa = "dipirona"

url = 'https://www.saojoaofarmacias.com.br/'+termo_de_pesquisa+'?_q='+termo_de_pesquisa+'&map=ft'

print(url)

headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 \
    (KHTML, like Gecko) Chrome / 86.0.4240.198Safari / 537.36"}

site = requests.get(url, headers=headers)
soup = BeautifulSoup(site.content, 'html.parser')
qtd_itens_div = 0
qtd_itens_div = soup.find_all('div', class_=re.compile('totalProducts'))
imprime_lista_produtos = pd.DataFrame
#print(qtd_itens_div)

if len(qtd_itens_div) > 0:
  qtd_itens_pesquisa = qtd_itens_div[0].get_text().split(" ")[0]

  qtd_itens_pesquisa

  dic_produtos = {'marca':[], 'preco':[]}

  url_pag = url
      
      
  produtos = soup.find_all('section', class_=re.compile('product-summary'))

  for produto in produtos:
      marca = produto.find('span', class_=re.compile('brandName')).get_text().strip()
      preco = produto.find('span', class_=re.compile('currencyInteger')).get_text().strip()

      #print(marca, preco)

      dic_produtos['marca'].append(marca)
      dic_produtos['preco'].append(int(preco))

  lista_produtos = pd.DataFrame(dic_produtos)

  index_drop = lista_produtos['marca'].str.contains(pat = termo_de_pesquisa, case=False)
  #index_drop = lista_produtos['marca'].str.contains(pat = termo_de_pesquisa+' 750Mg 20 comprimidos', case=False)  

  #print(index_drop)
  count = 0
  for index in index_drop:
    #print(index)
    if index==False:
      lista_produtos = lista_produtos.drop(count)
    
    count = count+1

  lista_produtos = lista_produtos.sort_values(by=['preco'], ascending=True)
```
